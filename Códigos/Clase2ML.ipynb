```python
import pandas as pd  #  Para trabajar con los datos 
import numpy as np # Operaciones 
import matplotlib.pyplot as plt # Hacer gr치ficos
import seaborn as sns # Hacer gr치ficos
from sklearn.datasets.samples_generator import make_blobs
```

```python
X,y = make_blobs(n_samples = 1000, centers = 2 , cluster_std = 2.8, random_state=2021)
plt.scatter(X[:,0],X[:,1], c=y)
plt.title('Soy una Gr치fica muy feliz !!!=)\n y ahora traigo datos de clusters')
```

```python
from sklearn.linear_model import SGDClassifier
clasificador = SGDClassifier(loss='hinge', alpha=0.1, max_iter=100, tol=0.01)
print('-'*64)
print()
print("Hola soy tu SVM !!! hicimos lo siguiente =)")
print()
print(clasificador.fit(X,y))
print()
print('-'*64)
```

```python
x_min , x_max = X[:,0].min() -0.5,X[:,0].max() + 0.5
y_min , y_max = X[:,1].min() -0.5,X[:,1].max() + 0.5
xx,yy = np.meshgrid(np.arange(x_min,x_max,0.05), np.arange(y_min,y_max,0.05))
```

```python
## Aplicar mi modelito
mi_clf =clasificador.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
```

```python
plt.contour(xx,yy,mi_clf)
plt.scatter(X[:,0],X[:,1], c=y)
plt.title('Soy tu primer modelo de ML y estoy muy feliz!!')
```

### Notas de clase 



```python
import pandas as pd  #  Para trabajar con los datos 
import numpy as np # Operaciones 
import matplotlib.pyplot as plt # Hacer gr치ficos
import seaborn as sns # Hacer gr치ficos
from sklearn.datasets.samples_generator import make_blobs
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_blobs(n_samples = 1000, centers = 2 , cluster_std = 2.8, random_state=2021)
plt.scatter(X[:,0],X[:,1], c=y)
plt.title('Soy una Gr치fica muy feliz !!!=)\n y ahora traigo datos de clusters')
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1], c=y)
```

```python
clasificador = SGDClassifier(loss='hinge', alpha=0.1, max_iter=100, tol=0.01)
print('-'*64)
print()
print("Hola soy tu SVM !!! hicimos lo siguiente =)")
print()
print(clasificador.fit(X,y))
print()
print('-'*64)
```

```python
x_min , x_max = X[:,0].min() -0.5,X[:,0].max() + 0.5
y_min , y_max = X[:,1].min() -0.5,X[:,1].max() + 0.5
xx,yy = np.meshgrid(np.arange(x_min,x_max,0.05), np.arange(y_min,y_max,0.05))
```

```python
## Aplicar mi modelito
mi_clf =clasificador.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
```

```python
plt.contour(xx,yy,mi_clf)
plt.scatter(X[:,0],X[:,1], c=y)
plt.title('Soy tu primer modelo de ML y estoy muy feliz!!')
```

## Atacar un problema de regresi칩n 

```python
a = .5
b = 1

x=20 * np.random.random(20)
y = a*x + b + np.random.normal(size=x.shape)
```

```python
plt.scatter(x,y)
```

```python
modelito_lineal = LinearRegression()
modelito_lineal.fit(x[:,None], y)
```

```python
print(modelito_lineal.coef_)
print(modelito_lineal.intercept_)
```

```python
x_new = np.linspace(0,30,100)
y_new = modelito_lineal.predict(x_new[:,None])
plt.scatter(x,y)
plt.plot(x_new,y_new,'g-')
plt.title("Soy tu regresi칩n Lineal! ")
```

## Una intro formal al Machine Learnig 

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.keys()
```

```python
## Convertir en una base de datos

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['Species'] = iris.target
iris_df['Species'] =iris_df['Species'].astype('category')
iris_df['Species'].cat.categories = iris.target_names
```

```python
iris_df
```

```python
sns.pairplot(iris_df,hue='Species')
```

```python
X,y = iris.data, iris.target
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(X)
X_reducida = pca.transform(X)
```

```python
plt.scatter(X_reducida[:,0],X_reducida[:,1], c=y)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1], c=y)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_reducida[:,0],X_reducida[:,1], c=y)
```

```python
from sklearn.manifold import Isomap
pca = Isomap(n_components=3)
pca.fit(X)
X_reducida2 = pca.transform(X)
```

```python
plt.scatter(X_reducida2[:,0],X_reducida2[:,1],c=y)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_reducida2[:,0],X_reducida2[:,1], c=y)
```
